```python
# Candidate Elimination Algorithm

def candidate_elimination(concepts, target):
    # Initialize S and G
    specific_hypothesis = concepts[0].copy()
    general_hypothesis = [["?" for _ in range(len(specific_hypothesis))]
                          for _ in range(len(specific_hypothesis))]

    print("Initial S:", specific_hypothesis)
    print("Initial G:", general_hypothesis)
    print("-" * 50)

    for i, example in enumerate(concepts):
        print(f"Training example {i + 1}: {example}, Target = {target[i]}")

        if target[i] == "Yes":  # Positive example
            for j in range(len(specific_hypothesis)):
                if example[j] != specific_hypothesis[j]:
                    specific_hypothesis[j] = "?"
                    general_hypothesis[j][j] = "?"
        else:  # Negative example
            for j in range(len(specific_hypothesis)):
                if example[j] != specific_hypothesis[j]:
                    general_hypothesis[j][j] = specific_hypothesis[j]
                else:
                    general_hypothesis[j][j] = "?"

        print("S:", specific_hypothesis)
        print("G:", general_hypothesis)
        print("-" * 50)

    # Remove overly general hypotheses
    general_hypothesis = [g for g in general_hypothesis if g != ["?" for _ in range(len(specific_hypothesis))]]

    return specific_hypothesis, general_hypothesis


# Example dataset
concepts = [
    ["Sunny", "Warm", "Normal", "Strong", "Warm", "Same"],
    ["Sunny", "Warm", "High", "Strong", "Warm", "Same"],
    ["Rainy", "Cold", "High", "Strong", "Warm", "Change"],
    ["Sunny", "Warm", "High", "Strong", "Cool", "Change"]
]

target = ["Yes", "Yes", "No", "Yes"]

# Run algorithm
S, G = candidate_elimination(concepts, target)

print("Final Specific Hypothesis:", S)
print("Final General Hypothesis:", G)
```
